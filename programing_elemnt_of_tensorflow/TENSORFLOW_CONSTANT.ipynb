```python
import tensorflow as tf
```

```python
# int contant
a=tf.constant(10)
a
```

```python
# float constant
b=tf.constant(20.0)
b
```

```python
# string constant
c=tf.constant("hello")
c
```

```python
# create numpy array
import numpy as np
```

```python
numpy_arr=tf.constant(np.array([1,2,3,4,5]))
numpy_arr
```

```python
# create 2d aray
numpy_arr=tf.constant(np.array([[1,2,3,4,5],[1,3,4,6,7]]))
numpy_arr
```

```python
# sahpe of 2d aray
numpy_arr=tf.constant([1,2,3,4],shape=(2,2))
numpy_arr
```

```python
# acces constant value
numpy_arr.numpy()
```
