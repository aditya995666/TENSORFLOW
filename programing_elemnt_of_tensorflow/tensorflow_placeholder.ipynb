```python
# create tensorfloq
import tensorflow as tf
```

```python
# execution compat.v1.disable_eager_execution
tf.compat.v1.disable_eager_execution()
```

```python
# create placeholed
a=tf.compat.v1.placeholder(dtype=tf.float32,shape=(400,400))

a
```

```python
b=tf.compat.v1.placeholder(dtype=tf.float32,shape=(400,400))
b
```

# ***add***

```python
c=tf.add(a,b)
c
```

# create numpy array

```python
import numpy as np
num_arr=np.ones((400,400),np.float32)
num_arr
```

execution tensorflow placeholder using session
## New Section

```python
with tf.compat.v1.Session() as sess:
  print(sess.run(c,feed_dict={a:num_arr,b:num_arr}))
```
