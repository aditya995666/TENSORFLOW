```python
# check name
a.name , a.shape , a.dtype
```

```python
# create list
b=tf.Variable([1,2,3])
b
```

```python
# create string
c=tf.Variable('hello')
c
```

```python
# create int,float
d=tf.Variable([1.2,6,1.3,45.4])
d
```

```python
d.numpy()
```

create variable from **constant**

```python
tf_constan=tf.constant([1,2,3])
tf_constan
```

```python
tf_var=tf.Variable(tf_constan)
tf_var
```

# create multiple matrix

```python
tf_2d=tf.Variable([[2,3,4,5],[4,5,6,7]],dtype="float32",shape=(2,4))
tf_2d
```

```python
tf.reshape(tf_2d,[4,2])
```

# **ccheck the shape**

```python
tf.reshape(tf_2d,(2,4))
```

# ***convert as a tensor***

```python
tf_1=tf.Variable([1,2,3])
tf_1
```

```python
m=tf.convert_to_tensor(tf_1)
```

# ***assign a new value***

```python
# Create a variable
r = tf.Variable([1, 2, 3], dtype=tf.int32)

# Use the assign method to update its value
r.assign([4, 5, 6])
```
